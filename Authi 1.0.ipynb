# Task
Create an AI model called Authi 1.0 that maps medical conditions from "Endocrine & Metabolic Disorders.csv" to their diagnostic and ongoing management treatment baskets in "Endocrine & Metabolic Disorders Treatment.csv", ensuring retrieval of structured treatment information aligned with Prescribed Minimum Benefits when a condition is queried.

## Load data

### Subtask:
Load the two CSV files ("Endocrine & Metabolic Disorders.csv" and "Endocrine & Metabolic Disorders Treatment.csv") into pandas DataFrames.


**Reasoning**:
Import pandas and load the two CSV files into dataframes as instructed.



```python
import pandas as pd

disorders_df = pd.read_csv('Authi-endocrine-dataset-training/Endocrine & Metabolic Disorders.csv')
treatment_df = pd.read_csv('Authi-endocrine-dataset-training/Endocrine & Metabolic Disorders Treatment.csv')

print("Disorders DataFrame head:")
display(disorders_df.head())

print("\nTreatment DataFrame head:")
display(treatment_df.head())
```

## Understand data structure

### Subtask:
Examine the columns and data types of both DataFrames to understand their structure and how they relate to each other.


**Reasoning**:
Display the column names and data types for both dataframes and the first few rows again to understand their structure and identify potential linking columns, and note any inconsistencies.



```python
print("Disorders DataFrame Info:")
disorders_df.info()

print("\nTreatment DataFrame Info:")
treatment_df.info()

print("\nDisorders DataFrame head:")
display(disorders_df.head())

print("\nTreatment DataFrame head:")
display(treatment_df.head())
```

## Merge data

### Subtask:
Merge the two DataFrames based on the medical condition to link each disorder with its diagnostic and ongoing management treatment baskets.


**Reasoning**:
Remove the extraneous header row from `treatment_df` and rename the columns as specified in the instructions to prepare for merging.



```python
treatment_df = treatment_df.iloc[1:].copy()
treatment_df.columns = [
    'CONDITION',
    'Diagnostic Procedure Description',
    'Diagnostic Procedure Code',
    'Number of Diagnostic Procedures',
    'Ongoing Management Description',
    'Ongoing Management Code',
    'Number of Ongoing Management Procedures',
    'Number of Specialists'
]
print("Cleaned Treatment DataFrame head:")
display(treatment_df.head())
```

**Reasoning**:
Perform a left merge between `disorders_df` and the cleaned `treatment_df` on the medical condition columns and store the result in a new DataFrame.



```python
merged_df = pd.merge(disorders_df, treatment_df, left_on='CHRONIC CONDITIONS', right_on='CONDITION', how='left')
print("Merged DataFrame head:")
display(merged_df.head())
```

## Structure knowledge base

### Subtask:
Organize the merged data into a structured format (e.g., a dictionary or JSON object) where each medical condition is a key, and its value contains the diagnostic and ongoing management baskets as nested structures.


**Reasoning**:
Define a function to structure the treatment information for a single condition and then iterate through the merged DataFrame to build the knowledge base dictionary.



```python
def structure_treatment_info(row):
    """Structures diagnostic and ongoing management information for a condition."""
    diagnostic_procedures = []
    if pd.notna(row['Diagnostic Procedure Description']):
        diagnostic_procedures.append({
            "Description": row['Diagnostic Procedure Description'],
            "Code": row['Diagnostic Procedure Code'],
            "Number of Procedures": row['Number of Diagnostic Procedures']
        })

    ongoing_management = []
    if pd.notna(row['Ongoing Management Description']):
        ongoing_management.append({
            "Description": row['Ongoing Management Description'],
            "Code": row['Ongoing Management Code'],
            "Number of Procedures": row['Number of Ongoing Management Procedures'],
            "Number of Specialists": row['Number of Specialists']
        })

    return {
        "Diagnostic Procedures": diagnostic_procedures,
        "Ongoing Management": ongoing_management
    }

knowledge_base = {}
for index, row in merged_df.iterrows():
    condition = row['CHRONIC CONDITIONS']
    treatment_info = structure_treatment_info(row)

    if condition not in knowledge_base:
        knowledge_base[condition] = {
            "Diagnostic Procedures": [],
            "Ongoing Management": []
        }

    knowledge_base[condition]["Diagnostic Procedures"].extend(treatment_info["Diagnostic Procedures"])
    knowledge_base[condition]["Ongoing Management"].extend(treatment_info["Ongoing Management"])

# Display a sample of the knowledge base
sample_conditions = list(knowledge_base.keys())[:5]
for condition in sample_conditions:
    print(f"Condition: {condition}")
    display(knowledge_base[condition])
```

## Prepare for retrieval

### Subtask:
Format the structured data in a way that can be easily queried and presented by an AI model.


**Reasoning**:
Iterate through the knowledge_base to remove duplicate entries within the diagnostic procedures and ongoing management lists for each condition to ensure unique treatment entries.



```python
def remove_duplicate_treatments(treatment_list):
    """Removes duplicate dictionaries from a list of treatment dictionaries."""
    unique_treatments = []
    seen = set()
    for treatment in treatment_list:
        # Convert dictionary to a frozenset of key-value pairs for hashing
        # Handle potential NaN values by converting them to a string representation
        treatment_frozenset = frozenset((k, str(v) if pd.isna(v) else v) for k, v in treatment.items())
        if treatment_frozenset not in seen:
            seen.add(treatment_frozenset)
            unique_treatments.append(treatment)
    return unique_treatments

for condition, treatments in knowledge_base.items():
    knowledge_base[condition]["Diagnostic Procedures"] = remove_duplicate_treatments(treatments["Diagnostic Procedures"])
    knowledge_base[condition]["Ongoing Management"] = remove_duplicate_treatments(treatments["Ongoing Management"])

# Display a sample of the refined knowledge base
sample_conditions = list(knowledge_base.keys())[:5]
for condition in sample_conditions:
    print(f"Condition: {condition}")
    display(knowledge_base[condition])
```
